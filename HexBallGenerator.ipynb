### Hexagon Ball Generator

Take a bunch of hexagons (of increasing then decreasing side) and stack them on top of each other<br>
making sure each layer is staggered from the last. Join the corners. This is what you get.

```python
from math import sin,cos,pi
```

```python
radius = 8 # ive only tried even numbers lol
```

```python
points = []
edges = []

for i in range(2*radius + 1):
    # POINTS
    z = radius - i
    if i==0:
        points.append([0,0,z])
    if i== 2*radius:
        points.append([0,0,z])
    rr = (radius**2 - abs(i-radius)**2)**0.5
    if i not in [0,2*radius]:
        if i%2==1:
            for ang in [90,150,210,270,330,30]:
                xc = 0 if abs(rr*cos(ang*pi/180))<10**-6 else rr*cos(ang*pi/180)
                yc = 0 if abs(rr*sin(ang*pi/180))<10**-6 else rr*sin(ang*pi/180)
                points.append([xc,yc,z])
        if i%2==0:
            for ang in [120,180,240,300,360,60]:
                xc = 0 if abs(rr*cos(ang*pi/180))<10**-6 else rr*cos(ang*pi/180)
                yc = 0 if abs(rr*sin(ang*pi/180))<10**-6 else rr*sin(ang*pi/180)
                points.append([xc,yc,z])
    
    if i==0: # top tent
        for j in range(6):
            edges.append([i,j+1])
    if i==2*radius: # bottom tent
        for j in range(6):
            edges.append([6*(i-1)+1,6*(i-2)+1+j])
    if i not in [0,2*radius]: # rings
        for j in range(6):
            edges.append([i*6-6 + j+1,(i*6-6 + j+2)- (6 if j == 5 else 0)])

    if i not in [0,2*radius,2*radius-1]:
        if i%2 ==1 :
            for j in range(1,6):
                edges.append([i*6-6 + j+1,i*6-6 + j+1+5])
                edges.append([i*6-6 + j+1,i*6-6 + j+1+6])
            edges.append([i*6-6 +1,i*6-6 +1+11])
            edges.append([i*6-6 +1,i*6-6 +1+6])
        if i%2 ==0 :
            for j in range(1,6):
                edges.append([i*6-6 + j,i*6-6 + j+6])
                edges.append([i*6-6 + j,i*6-6 + j+7])
            edges.append([i*6 ,i*6 + 1])
            edges.append([i*6 ,i*6 + 6])
```

```python
print("{edges:"+str(edges)+",points:"+str(points))
```
